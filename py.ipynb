```python
import cv2
from ultralytics import YOLO

# Load the YOLOv8 model
model = YOLO('yolov8n-pose.pt')

# 동영상 파일 사용시
# video_path = "path/to/your/video/file.mp4"
# cap = cv2.VideoCapture(video_path)

# webcam 사용시
#cap = cv2.VideoCapture(0)

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLOv8 inference on the frame
        results = model(frame)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        # Display the annotated frame
        cv2.imshow("YOLOv8 Inference", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import torch

# YOLOv8 모델 로드
model = YOLO('yolov8n-pose.pt')  # 모델 버전에 따라 변경 가능

# 웹캠 설정
cap = cv2.VideoCapture(0)  # 0은 기본 웹캠을 의미함. 다른 카메라를 사용할 경우 숫자 변경

while True:
    # 웹캠에서 프레임을 읽음
    ret, frame = cap.read()
    if not ret:
        break

    # YOLOv8에 이미지를 전달하여 탐지 수행
    results = model(frame)

    # 탐지 결과를 이미지에 그림
    results.render()

    # 결과 이미지를 화면에 표시
    cv2.imshow('YOLOv8 Webcam', results.imgs[0])

    # 'q'를 누르면 반복문을 종료하고 창을 닫음
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 자원 해제
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import torch

# YOLOv8 모델 로드
model = YOLO('yolov8n-pose.pt')  # 모델 버전에 따라 변경 가능

# 웹캠 설정
cap = cv2.VideoCapture(0)  # 0은 기본 웹캠을 의미함. 다른 카메라를 사용할 경우 숫자 변경

while True:
    # 웹캠에서 프레임을 읽음
    ret, frame = cap.read()
    if not ret:
        break

    # 이미지에 대한 객체 탐지 수행
    results = model(frame)

    # 탐지 결과를 이미지에 그림 (수정된 부분)
    results.render()  # results 객체에 직접 .render() 호출

    # 결과 이미지를 화면에 표시
    cv2.imshow('YOLOv8 Webcam', results.imgs[0] if hasattr(results, 'imgs') else results.render()[0])

    # 'q'를 누르면 반복문을 종료하고 창을 닫음
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 자원 해제
cap.release()
cv2.destroyAllWindows()
```
